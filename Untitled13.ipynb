```python
import gradio as gr
import cv2
import torch
import numpy as np
import easyocr  # Improved OCR
from ultralytics import YOLO
from twilio.rest import Client

# ‚úÖ Load YOLO models
vehicle_model = YOLO("yolov8n.pt")  # Detects vehicles
plate_model = YOLO("yolov8n.pt")  # Using same YOLO model (may not be perfect for plates)

# ‚úÖ Twilio API credentials (Replace with actual credentials)
TWILIO_SID = "AC50091e7238247f58f4f9acb3c3d88975"
TWILIO_AUTH_TOKEN = "ed958463f0ae45f48423051ac0c1ce00"
TWILIO_PHONE_NUMBER = "+19897188207"
POLICE_PHONE_NUMBER = "+918056414747"

# ‚úÖ Load OCR (EasyOCR is better than Tesseract)
ocr_reader = easyocr.Reader(["en"])  # English OCR

def send_alert(plate_number):
    """Send SMS alert to police with vehicle number."""
    try:
        client = Client(TWILIO_SID, TWILIO_AUTH_TOKEN)
        message = client.messages.create(
            body=f"üö® Accident detected! Vehicle Number: {plate_number}.",
            from_=TWILIO_PHONE_NUMBER,
            to=POLICE_PHONE_NUMBER
        )
        print("‚úÖ Alert Sent:", message.sid)
        return f"Alert sent: {plate_number}"
    except Exception as e:
        print(f"‚ö† Error sending SMS: {e}")
        return f"Failed to send alert: {e}"

def extract_number_plate(frame, results):
    """Detect and extract number plate text using YOLO & OCR."""
    for result in results:
        for box in result.boxes:
            class_id = int(box.cls)
            confidence = float(box.conf)

            if class_id == 0 and confidence > 0.6:  # Assuming class 0 = number plate
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                plate_img = frame[y1:y2, x1:x2]

                # ‚úÖ Preprocess plate image (Better OCR)
                gray = cv2.cvtColor(plate_img, cv2.COLOR_BGR2GRAY)
                gray = cv2.bilateralFilter(gray, 11, 17, 17)  # Noise reduction
                _, binary = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

                # ‚úÖ OCR for text extraction
                plate_text = ocr_reader.readtext(binary, detail=0)
                plate_number = " ".join(plate_text).strip()

                return plate_img, plate_number if plate_number else "Not detected"
    return None, "Not detected"

def process_video(video_path):
    """Process video: detect accidents & extract number plates."""
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return "‚ö† Error: Cannot open video file!", None, "No plate detected."

    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    out = cv2.VideoWriter("output.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (frame_width, frame_height))

    accident_detected = False
    plate_number = "Not detected"
    alert_message = "‚úÖ No accident detected."

    for frame_id in range(0, total_frames, 5):  # Process every 5th frame for efficiency
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_id)
        ret, frame = cap.read()
        if not ret:
            break

        # ‚úÖ Detect Vehicles
        results = vehicle_model(frame)
        num_vehicles = len(results[0].boxes)

        # ‚úÖ Check for accidents
        for result in results:
            for box in result.boxes:
                class_id = int(box.cls)
                confidence = box.conf
                if class_id in [2, 3, 5, 7] and confidence > 0.6:  # 2=car, 3=motorcycle, 5=bus, 7=truck
                    for other_box in result.boxes:
                        if other_box != box:
                            ox1, oy1, ox2, oy2 = map(int, other_box.xyxy[0])
                            x1, y1, x2, y2 = map(int, box.xyxy[0])

                            if (x1 < ox2 and x2 > ox1 and y1 < oy2 and y2 > oy1):
                                accident_detected = True

                                # ‚úÖ Detect number plate
                                plate_results = plate_model(frame)
                                plate_img, plate_number = extract_number_plate(frame, plate_results)

                                # ‚úÖ Send alert if accident detected
                                if plate_number != "Not detected":
                                    alert_message = send_alert(plate_number)

                                # ‚úÖ Draw accident alert
                                cv2.rectangle(frame, (x1, oy1), (ox2, oy2), (0, 0, 255), 3)
                                cv2.putText(frame, "‚ö† CRASH", (x1, oy1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        # ‚úÖ Annotate frame
        annotated_frame = results[0].plot()
        out.write(annotated_frame)

    cap.release()
    out.release()

    return "output.mp4", alert_message

# ‚úÖ Gradio UI
iface = gr.Interface(
    fn=process_video,
    inputs=gr.Video(),
    outputs=[gr.Video(), gr.Text()],
    title="Accident & Number Plate Detection",
    description="Upload a video to detect accidents, extract vehicle number plates, and send to police."
)

iface.launch(share=True)
```
